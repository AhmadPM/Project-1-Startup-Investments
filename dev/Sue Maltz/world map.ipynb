```python
import pandas as pd
import numpy as np
import requests
import json
import gmaps

from api_keys import gkey

# Access maps with unique API key
gmaps.configure(api_key=gkey)
```

```python
file="full_map.csv"

df1=pd.read_csv(file)
df1
```

```python
df1["Lat"]=""
df1["Lng"]=""
```

```python
# create a params dict that will be updated with new city each iteration
params = {"key": gkey}

# Loop through the cities_pd and run a lat/long search for each city
for index, row in df1.iterrows():
    url = "https://maps.googleapis.com/maps/api/geocode/json"
    country = row['Country']
    
    # update address key value
    params['address'] = f"{country}"
    
    # make request
    df1_lat_lng = requests.get(url, params=params)
    
    # convert to json
    df1_lat_lng = df1_lat_lng.json()
    
    df1.loc[index, "Lat"] = df1_lat_lng["results"][0]["geometry"]["location"]["lat"]
    df1.loc[index, "Lng"] = df1_lat_lng["results"][0]["geometry"]["location"]["lng"]
    
# Visualize to confirm lat lng appear
df1.head()
```

```python
#Get the coordinates to use in gmaps
coordinate = df1[['Lat', 'Lng']]
coordinate
```

```python
fig = gmaps.figure()

# Assign the marker layer to a variable
markers = gmaps.marker_layer(coordinate)
# Add the layer to the map
fig.add_layer(markers)
fig
```
